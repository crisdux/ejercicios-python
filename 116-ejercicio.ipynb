```python
#  Dada una lista de valores numéricos, encuentre el valor máximo y el valor mínimo de la misma. No se pueden utilizar las funciones ││ "built-in" max() y min(), y tampoco sorted().  
```

```python
def max_value(values):
    max_value = values[0]
    for value in values:
        if value > max_value:
            max_value = value
    return max_value

# print(max_value( [4, 6, 2, 1, 9, 63, -134, 566] ))

def min_value(values):
    min_value = values[0]
    for value in values:
        if value < min_value:
            min_value = value
    return min_value


# print(min_value([4, 6, 2, 1, 9, 63, -134, 566]))

def maxmin_values(values):
    max_v = max_value(values)
    min_v= min_value(values)
    return max_v, min_v

print(maxmin_values([4, 6, 2, 1, 9, 63, -134, 566]))
print(maxmin_values([-52, 56, 30, 29, -54, 0, -110]))
print(maxmin_values( [42, 54, 65, 87, 0]))
print(maxmin_values([5]))
```
